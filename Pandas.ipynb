# Merge Dataframes

### Inner merge

#### Merge on single column

```python
# Creating the first DataFrame
data1 = {
    'FellowshipID': [101, 102, 103, 104, 105],
    'FirstName': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Skills': ['Python', 'Java', 'C++', 'Python', 'Java']
}

df1 = pd.DataFrame(data1)
df1
```

```python
# Creating the second DataFrame
data2 = {
    'FellowshipID': [101, 102, 105, 106, 107],
    'FirstName': ['Alice', 'Bob', 'Emma', 'Frank', 'Grace'],
    'Age': [25, 30, 28, 35, 29]
}

df2 = pd.DataFrame(data2)
df2
```

```python
df1.merge(df2, how='inner', on='FellowshipID') #shows common rows
```

#### Merge on multiple columns

```python
df1.merge(df2, how='inner', on=['FellowshipID', 'FirstName']) #shows common rows
```

### Outer merge

```python
df1.merge(df2, how='outer') #all rows
```

### Left merge

```python
df1.merge(df2, how='left')
```

### Right merge

```python
df1.merge(df2, how='right')
```

### Cross merge

```python
df1.merge(df2, how='cross')
```

## Join Dataframes

```python

# Dont use it this way. Output is less understandable
df1.join(df2, on='FellowshipID', how='outer', lsuffix='_Left', rsuffix='_Right')
```

```python
df3 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix='_Left', rsuffix='_Right')
df3
```

## Concatenate Dataframes

### Concatenate with all columns

```python
pd.concat([df1, df2])
```

### Concatenate as well as inner join

```python
pd.concat([df1, df2], join='inner')
```

### Concatenate as well as outer join

#### Stack rows on top of each other

```python
pd.concat([df1, df2], join='outer', axis=0)
```

#### Stack columns on top of each other

```python
pd.concat([df1, df2], join='outer', axis=1)
```

```python
df = pd.read_excel('/content/drive/MyDrive/GenAI/Pandas/data/Customer Call List.xlsx')
df.head()
```

## Iterate on all rows of a DataFrame

```python
for x in df.index:
    print(df.iloc[x]) #since iloc works on integer index
```

## Drop duplicates

```python
df = df.drop_duplicates()
df
```

## Drop Rows

```python
df = df.drop([14,15]) #14 and 15 are index
df
```

## Drop columns

```python
df = df.drop(columns = ["Not_Useful_Column"])
df
```

## Left/Right strip characters from particular column data

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df
```

## Left/Right string using regex

```python
df = pd.read_excel('/content/drive/MyDrive/GenAI/Pandas/data/Customer Call List.xlsx')
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip("123._/")
df['Last_Name']
```

## Lambda

### Convert phone numbers to this format 123-456-7890

```python
# First clean the phone numbers column
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '') #everything except a-z, A-Z, 9-9
df
```

```python
# Another is using lambda
# First convert phone number column to string
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))

# Then
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:10])
df['Phone_Number']
```

## Remove nan-- or NA-- from phone numbers

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df['Phone_Number']
```

## Splitting

```python
df['Address'].str.split(',', 2)
```

### See result in columns properly

```python
df['Address'].str.split(',', 2, expand=True)
```

### Assign the result to street address, state and zip code

```python
df[["Street_Address", "State", "Zip_Code"]] = df['Address'].str.split(',', 2, expand=True)
```

```python
df
```

## Replace

### Replace Yes/No with Y/N

```python
df["Paying Customer"] = df["Paying Customer"].str.replace("Yes", "Y")
df["Paying Customer"] = df["Paying Customer"].str.replace("No", "N")

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes", "Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No", "N")
df
```

### Replace N/a or NAN with empty

```python
df = df.replace('N/a', '')
df = df.fillna('')
df
```

## Drop columns based on condition

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"]=='Y':
        df.drop(x, inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"]=='':
        df.drop(x, inplace=True)
df
```

## Another way to drop null values from Phone number column

```python
df = df.dropna(subset="Phone_Number")
```

## Reset index finally

```python
df = df.reset_index(drop=True)
df
```
